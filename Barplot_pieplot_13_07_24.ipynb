```python
# Descriptive statistics
# Barplot is used for data visualization, 
#we can plot categorical verses numerical
# Barplot is a type of data visualization used to represent 
#data in the form of rectangular bars
# Height each bar represents the value of datapoint
# plt is alias name
# Numpy*(Numerical Python) is python library for 
#multidimensional arrays
# np.array it converts list to array
#.pyplot is a collection command style functions that make
#matplotlib work like a MATLAB,pyplot is a submodule
# Matplotlib data visualization
#import matplotlib.pyplot as plt
#import numpy as np
x=np.array(["A","B","c","D"])
# Categorical 
y=np.array([3,8,1,10])
plt.xlabel(" Categorical values")
plt.ylabel(" Numerical Values")
# Numerical
plt.title("Bargraph plot")
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["A","B","c","D"]
y=[3,8,1,10]
plt.bar(x,y)
plt.show()
```

```python
x=np.array(["A","B","c","D"])
# Categorical 
y=np.array([3,8,1,10])
# Numerical
plt.bar(x,y)
plt.show()
```

```python
# Add colors
x=np.array(["A","B","c","D"])
# Categorical 
y=np.array([3,8,1,10])
plt.xlabel(" Categorical values")
plt.ylabel(" Numerical Values")
# Numerical
plt.title("Bargraph plot")
plt.bar(x,y,color="red")
plt.show()
```

```python
# Matplotlib data visualization
#import matplotlib.pyplot as plt
#import numpy as np
x=np.array(["A","B","c","D"])
# Categorical 
y=np.array([3,8,1,10])
plt.xlabel("Numerical Values")
plt.ylabel("Categorical value")
# Numerical
plt.title("Bargraph plot")
plt.bar(x,y,color=["red","blue","hotpink","green"],width=0.2)
plt.show()
```

```python
# Horizontal Graph
x=np.array(["A","B","c","D"])
# Categorical 
y=np.array([3,8,1,10])
plt.xlabel(" Numerical Values")
plt.ylabel(" Categorical values")
# Numerical
plt.title("Bargraph plot")
plt.barh(x,y,color=["red","blue","blue","Green"],height=0.2)
plt.show()
```
